### 1. Загрузка и первичный анализ данных

```python
df = pd.read_csv('docs\PP12_ISP23_dashboard.csv')

print("=" * 60)
print("=== ЗАГРУЗКА ДАННЫХ ===")
print("=" * 60)
print(f"\nРазмер данных: {df.shape}")
print(f"\nПервые 5 записей:")
print(df.head())
```

```python
print("=" * 60)
print("=== ИНФОРМАЦИЯ О ДАННЫХ ===")
print("=" * 60)
print("\nТипы данных и количество непустых значений:")
print(df.info())

print("=" * 60)
print("=== СТАТИСТИЧЕСКОЕ ОПИСАНИЕ ===")
print("=" * 60)
print("\nОсновные статистики:")
print(df.describe())

print("=" * 60)
print("=== ПРОВЕРКА НА ПРОПУСКИ ===")
print("=" * 60)
print(f"\nПропущенные значения:\n{df.isnull().sum()}")
print(f"\nДубликаты: {df.duplicated().sum()}")

print("=" * 60)
print("=== УНИКАЛЬНЫЕ ЗНАЧЕНИЯ КАТЕГОРИЙ ===")
print("=" * 60)
print(f"\nКатегории: {df['category'].unique()}")
print(f"Регионы: {df['region'].unique()}")
print(f"Период данных: {df['date'].min()} — {df['date'].max()}")
```

### 2. Расчёт ключевых показателей (KPI)

```python
print("=" * 60)
print("=== КЛЮЧЕВЫЕ ПОКАЗАТЕЛИ (KPI) ===")
print("=" * 60)

total_sales = df['sales'].sum()
total_profit = df['profit'].sum()
avg_sales = df['sales'].mean()
avg_profit = df['profit'].mean()
profit_margin = (total_profit / total_sales) * 100

print(f"\n📊 ОБЩИЕ ПРОДАЖИ: ${total_sales:,.2f}")
print(f"💰 ОБЩАЯ ПРИБЫЛЬ: ${total_profit:,.2f}")
print(f"📈 СРЕДНИЕ ПРОДАЖИ: ${avg_sales:,.2f}")
print(f"💵 СРЕДНЯЯ ПРИБЫЛЬ: ${avg_profit:,.2f}")
print(f"📊 МАРЖА ПРИБЫЛИ: {profit_margin:.2f}%")

print("=" * 60)
print("=== KPI ПО КАТЕГОРИЯМ ===")
print("=" * 60)

category_stats = df.groupby('category').agg({
    'sales': ['sum', 'mean'],
    'profit': ['sum', 'mean']
}).round(2)

print("\n", category_stats)

print("=" * 60)
print("=== KPI ПО РЕГИОНАМ ===")
print("=" * 60)

region_stats = df.groupby('region').agg({
    'sales': ['sum', 'mean'],
    'profit': ['sum', 'mean']
}).round(2)

print("\n", region_stats)

print("=" * 60)
print("=== ТАБЛИЦА: ТОП-10 ПРОДАЖ ПО ДАТАМ ===")
print("=" * 60)

top_sales = df.nlargest(10, 'sales')[['date', 'category', 'region', 'sales', 'profit']]
print(top_sales.to_string(index=False))

# Дополнительно: сводная таблица
print("\n" + "=" * 60)
print("=== СВОДНАЯ ТАБЛИЦА: ПРОДАЖИ ПО КАТЕГОРИЯМ И РЕГИОНАМ ===")
print("=" * 60)

pivot_table = df.pivot_table(values='sales', 
                              index='category', 
                              columns='region', 
                              aggfunc='sum', 
                              margins=True,
                              margins_name='Итого')
print(pivot_table.round(2))
```

```python
print("=" * 60)
print("=== ПРОДАЖИ ПО РЕГИОНАМ И КАТЕГОРИЯМ ===")
print("=" * 60)

region_category = df.groupby(['region', 'category'])['sales'].sum().reset_index()

fig_grouped = px.bar(region_category, x='region', y='sales',
                     color='category',
                     title='Продажи по регионам и категориям',
                     labels={'region': 'Регион', 'sales': 'Продажи ($)', 'category': 'Категория'},
                     barmode='group')

fig_grouped.update_layout(
    width=1000,
    height=500,
    xaxis_title='Регион',
    yaxis_title='Продажи ($)',
    title_font_size=16
)

fig_grouped.show()

print("\n✅ Интерактивный график: кликните на легенду для фильтрации категорий")
```

### 4. Интерактивный дашборд с фильтрами

```python
print("=" * 60)
print("=== ИНТЕРАКТИВНЫЙ ДАШБОРД С ФИЛЬТРАМИ ===")
print("=" * 60)

# Преобразуем дату в datetime
df['date'] = pd.to_datetime(df['date'])

# Создаём фильтры
category_widget = widgets.Dropdown(
    options=['Все'] + sorted(df['category'].unique().tolist()),
    value='Все',
    description='Категория:',
    layout=widgets.Layout(width='30%')
)

region_widget = widgets.Dropdown(
    options=['Все'] + sorted(df['region'].unique().tolist()),
    value='Все',
    description='Регион:',
    layout=widgets.Layout(width='30%')
)

# DatePicker для выбора периода
date_start = widgets.DatePicker(
    description='Дата начала:',
    disabled=False,
    value=df['date'].min(),
    layout=widgets.Layout(width='30%')
)

date_end = widgets.DatePicker(
    description='Дата окончания:',
    disabled=False,
    value=df['date'].max(),
    layout=widgets.Layout(width='30%')
)

print("\nФильтры созданы. Ниже отображается дашборд...")
print("Изменяйте значения для обновления графиков.\n")

def update_dashboard(change=None):
    clear_output(wait=True)
    
    # Применяем фильтры
    mask = pd.Series(True, index=df.index)
    
    if category_widget.value != 'Все':
        mask &= (df['category'] == category_widget.value)
    if region_widget.value != 'Все':
        mask &= (df['region'] == region_widget.value)
    
    # Фильтр по периоду
    mask &= (df['date'] >= date_start.value)
    mask &= (df['date'] <= date_end.value)
    
    filtered = df[mask].copy()
    
    # Отображаем виджеты
    display(widgets.HBox([category_widget, region_widget]))
    display(widgets.HBox([date_start, date_end]))
    
    # KPI
    print("\n" + "=" * 60)
    print("=== КЛЮЧЕВЫЕ ПОКАЗАТЕЛИ (KPI) ===")
    print("=" * 60)
    print(f"Общие продажи: ${filtered['sales'].sum():,.2f}")
    print(f"Общая прибыль: ${filtered['profit'].sum():,.2f}")
    print(f"Средняя прибыль: ${filtered['profit'].mean():,.2f}")
    print(f"Записей: {len(filtered)}")
    
    # Таблица топ-5 продаж
    if len(filtered) > 0:
        print("\n" + "=" * 60)
        print("=== ТОП-5 ПРОДАЖ (ВЫБРАННЫЙ ПЕРИОД) ===")
        print("=" * 60)
        top5 = filtered.nlargest(5, 'sales')[['date', 'category', 'region', 'sales']]
        print(top5.to_string(index=False))
    
    # График 1: Динамика продаж
    if len(filtered) > 0:
        fig1 = px.line(filtered, x='date', y='sales', 
                       title='Динамика продаж',
                       markers=True)
        fig1.update_layout(width=900, height=400)
        fig1.show()
        
        # График 2: Продажи по категориям (Pie Chart!)
        if category_widget.value == 'Все':
            cat_sales = filtered.groupby('category')['sales'].sum().reset_index()
            fig2 = px.pie(cat_sales, values='sales', names='category',
                         title='Доля категорий в продажах',
                         hole=0.4)
            fig2.update_layout(width=900, height=400)
            fig2.show()
        
        # График 3: Продажи по регионам
        if region_widget.value == 'Все':
            reg_sales = filtered.groupby('region')['sales'].sum().reset_index()
            fig3 = px.bar(reg_sales, x='region', y='sales',
                         title='Продажи по регионам',
                         color='sales')
            fig3.update_layout(width=900, height=400)
            fig3.show()
    else:
        print("\n⚠️ Нет данных для выбранных фильтров!")

# Привязываем обработчики
category_widget.observe(update_dashboard, names='value')
region_widget.observe(update_dashboard, names='value')
date_start.observe(update_dashboard, names='value')
date_end.observe(update_dashboard, names='value')

# Первый запуск
update_dashboard()
```

### 5. Аналитика и выводы

```python
print("=" * 60)
print("=== АНАЛИЗ ТРЕНДОВ ===")
print("=" * 60)

# Динамика по месяцам
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.to_period('M')

monthly_sales = df.groupby('month')['sales'].sum()

best_month = monthly_sales.idxmax()
worst_month = monthly_sales.idxmin()

print(f"\n📈 Лучший месяц: {best_month} (${monthly_sales[best_month]:,.2f})")
print(f"📉 Худший месяц: {worst_month} (${monthly_sales[worst_month]:,.2f})")
print(f"📊 Разница: ${monthly_sales[best_month] - monthly_sales[worst_month]:,.2f}")

print("=" * 60)
print("=== ТОП КАТЕГОРИЙ И РЕГИОНОВ ===")
print("=" * 60)

# Топ категория
top_category = df.groupby('category')['sales'].sum().idxmax()
top_category_sales = df.groupby('category')['sales'].sum().max()

# Топ регион
top_region = df.groupby('region')['sales'].sum().idxmax()
top_region_sales = df.groupby('region')['sales'].sum().max()

print(f"\n🏆 Топ категория: {top_category} (${top_category_sales:,.2f})")
print(f"🏆 Топ регион: {top_region} (${top_region_sales:,.2f})")
```
